```python
mapping_df = pd.read_excel('../DAT/[YP2021] 1-2차 조사 변수매핑표.xlsx')
mapping_df.head()
```

```python
# 데이터 로드 및 컬럼명 재설정
mapping_df.columns = ['영역구분', '변수설명', 'w02', 'w01', '특이사항']
mapping_df = mapping_df.reset_index(drop=True)
# 데이터 확인
print(mapping_df.head())
```

```python
# 관심 키워드 정의
keywords = ['부동산', '소득', '임금', '취업', '학력', '종사상지위']


# 키워드 기반 필터링
filtered_vars = mapping_df[mapping_df['변수설명'].str.contains('|'.join(keywords), na=False)]

# 결과 확인
print(filtered_vars[['변수설명','w02','w01']])

```

```python
filtered_vars.to_csv('../DAT/filter.csv')
```

```python
# 필요한 변수 목록 추출
w01_vars = filtered_vars['w01'].dropna().tolist()
w02_vars = filtered_vars['w02'].dropna().tolist()

w01_vars = [var.strip().upper() for x in w01_vars for var in x.split('/')]
w02_vars = [x.upper() for x in w02_vars]
# df1과 df2에서 필요한 변수 선택
df1_selected = df1[['SAMPID'] + w01_vars]
df2_selected = df2[['SAMPID'] + w02_vars]

# 두 데이터 프레임을 SAMPID를 기준으로 병합
merged_df = pd.merge(df1_selected, df2_selected, on='SAMPID', how='outer')

# 결과 확인
print(merged_df.head())
```
